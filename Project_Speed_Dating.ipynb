```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset= pd.read_csv("Speed_Dating_Data.csv", encoding='ISO-8859-1')
```

```python
pd.set_option('display.max_columns', None)
dataset.head()
```

```python
dataset1= dataset[["iid","gender","wave","age","samerace","field","imprace","imprelig","race","art","shopping","attr1_1","sinc1_1","intel1_1","amb1_1","shar1_1","satis_2","match","dec_o"]]
```

```python
dataset1.columns
```

```python
dataset2 = dataset1.drop_duplicates(subset = "iid")
```

```python
dataset2.head()
```

```python
dataset2= dataset2.rename(columns={"attr1_1" : "Attractive",
                         "sinc1_1" : "Sincere",
                         "intel1_1" : "Intelligent",
                         "amb1_1" : "Ambitious",
                         "shar1_1" : "Shared Interests",
                         "satis_2" : "Satisfied",
                         "dec_o" : "Decision_Partner"})

dataset2
```

```python
dataset2["race"]=dataset2["race"].apply(lambda x:
                                    "Black/African American"      if x==1
                                else"European/Caucasion-American" if x==2
                               else"Latino/Hispanic American"     if x==3
                               else"Asian/Pacific Islander/Asian-American" if x==4
                               else"Native American"              if x==5
                               else "Other"                       if x==6 
                               else x)
```

```python

dataset2["gender"]=dataset2["gender"].apply(lambda x:
                                         "femele"   if x==1
                                     else"male"    if x==0
                                     else x)
```

```python
dataset2["match"]=dataset2["match"].apply(lambda x:
                                        "yes" if x==1
                                    else"no" if x==0 
                                    else x)
```

```python
dataset2["samerace"]=dataset2["samerace"].apply(lambda x:"yes"   if x==1 else"no"    if x==0 else x)
```

```python
dataset2.isnull().sum()
```

```python
dataset3 = dataset2.dropna()
dataset3.shape
```

```python
dataset3.head()
```

```python
dataset3["match"]
```

```python
sns.relplot(x="age", y="Decision_Partner", kind="line", data=dataset3)
```

```python
sns.histplot(data=dataset3, x="age", hue='Decision_Partner')
```

```python
def transform_age(age):
    if age > 40:
        return "> 40"
    elif age > 35 and age <= 40:
        return "> 35"
    else:
        return "< 35"
```

```python
dataset3['age_label'] = dataset3['age'].apply(lambda x: transform_age(x))
```

```python
sns.catplot(x="age_label", y="Decision_Partner", kind="bar", data=dataset3, hue='gender' )
```

```python
dataset3["Decision_Partner"]=dataset3["Decision_Partner"].apply(lambda x:
                                                  "yes"   if x==1
                                              else"no"    if x==0
                                               else x)
```

```python
sns.catplot(x="Satisfied", y="Decision_Partner", data = dataset3, kind= "box")
```
